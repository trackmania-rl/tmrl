```python
class Class1:
    def __init__(self):
        print(f"init 1")
    
    def step_internal(self):
        print("internal 1")
    
    def step(self):
        print("step 1")
        self.step_internal()


class Class2(Class1):
    def __init__(self):
        print(f"init 2")
        super(Class2, self).__init__()
    
    def step_internal(self):
        print("internal 2")
        super().step_internal()

class Class3(Class2):
    def __init__(self):
        print(f"init 3")
        super(Class3, self).__init__()
    
    def step_internal(self):
        print("internal 3")
    
    def step(self):
        print("step 3")
        super().step()
```

```python
c3 = Class3()
```

```python
c3.step()
```
